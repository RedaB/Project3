```python
#dependencies 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
from matplotlib.pyplot import rcParams 
import warnings                   # To ignore the warnings
warnings.filterwarnings("ignore")
rcParams['figure.figsize'] = 10,6 
```

```python
# Read the csv file into a pandas DataFrame
df = pd.read_csv('/Users/benjaminhuypham/Desktop/Energy_Project/Resources/AEP_hourly.csv')
df.head()
```

```python
#parse string to datetime type 
df['Datetime'] = pd.to_datetime(df['Datetime'], infer_datetime_format = True)
indexDf = df.set_index(['Datetime'])
```

```python
from datetime import datetime 
indexDf.head(5)
```

```python
#plot graph 
plt.xlabel('Date')
plt.ylabel('Energy Consumption by MW')
plt.plot(indexDf)
```

```python
#Determine rolling statistics 
rolMean = indexDf.rolling(window=12).mean()
rolStd = indexDf.rolling(window=12).std() 
# print(rolMean, rolStd)
```

```python
#plot rolling statistics 
orig = plt.plot(indexDf, c='blue', label='Original')
mean = plt.plot(rolMean, c='red', label='Rolling Mean')
std = plt.plot(rolStd, c='green', label='Rolling Std')
plt.legend(loc='Best')
plt.title("Rolling mean and standard deviation")
plt.show(block=False)
```

```python
#Perform Dickey-Fuller test: 
from statsmodels.tsa.stattools import adfuller 

print('Results of Dicky-Fuller Test:')
dftest = adfuller(indexDf['AEP_MW'], autolag='AIC') 
dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#Lags Used', 'Number of Observation Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value 
    
print(dfoutput)
```

```python
#Estimating Trend 
indexDf_logScale = np.log(indexDf)
plt.plot(indexDf_logScale)
```

```python
movingAverage = indexDf_logScale.rolling(window=12).mean()
movingStd = indexDf_logScale.rolling(window=12).std()
plt.plot(indexDf_logScale)
plt.plot(movingAverage, c='red')
```

```python
dfLogScaleMinusMovingAverage = indexDf_logScale - movingAverage 
dfLogScaleMinusMovingAverage.head()

#Remove NaN values 
dfLogScaleMinusMovingAverage.dropna(inplace=True)
dfLogScaleMinusMovingAverage.head()
```

```python
from statsmodels.tsa.stattools import adfuller 
def test_stationary(timeseries):
    
    #determine rolling statistics 
    movingAverage = timeseries.rolling(window=12).mean()
    movingStd = timeseries.rolling(window=12).std()
    
    #plot rolling statistics: 
    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(movingAverage, color='red', label='Rolling Mean')
    std = plt.plot(movingStd, color='green', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test: 
    print('Results of Dickey-Fuller Test:')
    dfTest = adfuller(timeseries['AEP_MW'], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistics','p-value','#Lag Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value 
    print(dfoutput)
    
```

```python
test_stationary(dfLogScaleMinusMovingAverage)
```

```python
exponentialDecayWeightedAverage = indexDf_logScale.ewm(halflife=12, min_periods=0, adjust=True).mean()
plt.plot(indexDf_logScale)
plt.plot(exponentialDecayWeightedAverage, c='red')
```

```python
dfLogScaleMinusMovingExponentialDecayWeightedAverage = indexDf_logScale - exponentialDecayWeightedAverage
test_stationary(dfLogScaleMinusMovingExponentialDecayWeightedAverage)
```

```python
dfLogDiffShifting = indexDf_logScale - indexDf_logScale.shift()
plt.plot(dfLogDiffShifting)
```

```python
dfLogDiffShifting.dropna(inplace=True)
test_stationary(dfLogDiffShifting)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(indexDf_logScale, model = 'additive', freq=1)

trend = decomposition.trend
seasonal = decomposition.seasonal 
residual = decomposition.resid 

plt.subplot(411)
plt.plot(indexDf_logScale, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()

decomposedLogData = residual 
decomposedLogData.dropna(inplace=True)
test_stationary(decomposedLogData)
```

```python
decomposedLogDf = residual 
decomposedLogDf.dropna(inplace=True)
test_stationary(decomposedLogDf)
```

```python
#ACF and PACF plots: 
from statsmodels.tsa.stattools import acf, pacf 

lag_acf = acf(dfLogDiffShifting, nlags=20)
lag_pacf = pacf(dfLogDiffShifting, nlags=20, method = 'ols')

#Plot ACF:
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(dfLogDiffShifting)), linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(dfLogDiffShifting)), linestyle='--', color='gray')
plt.title("Autocorrelation Function")

#Plot PACF: 
plt.subplot(121)
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(dfLogDiffShifting)), linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(dfLogDiffShifting)), linestyle='--', color='gray')
plt.title("Partial Autocorrelation Function")
plt.tight_layout()

```

```python
from statsmodels.tsa.arima_model import ARIMA

#AR Model 
model = ARIMA(indexDf_logScale, order=(2,1,2))
results_AR = model.fit(disp=1)
plt.plot(dfLogDiffShifting)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues - dfLogDiffShifting['AEP_MW'])**2))
print('Plotting AR Model')
```

```python
#MA Model 
model = ARIMA(indexDf_logScale, order=(0,1,2))
results_AR = model.fit(disp=1)
plt.plot(dfLogDiffShifting)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues - dfLogDiffShifting['AEP_MW'])**2))
print('Plotting MA Model')
```

```python
model = ARIMA(indexDf_logScale, order=(2,1,2))
results_ARIMA = model.fit(disp=1)
plt.plot(dfLogDiffShifting)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues - dfLogDiffShifting['AEP_MW'])**2))
print('Plotting MA Model')
```

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
print(predictions_ARIMA_diff.head())
```

```python
#convert to culmulative sum 
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print(predictions_ARIMA_diff_cumsum.head())
```

```python
predictions_ARIMA_log = pd.Series(indexDf_logScale['AEP_MW'].ix[0], index=indexDf_logScale.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum, fill_value=0)
predictions_ARIMA_log.head()
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(indexDf)
plt.plot(predictions_ARIMA)
```
